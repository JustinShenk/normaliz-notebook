## PyNormaliz

Calculate the Hilbert basis from finite vectors.

```python
%matplotlib notebook
import PyNormaliz
import numpy as np
import matplotlib.pyplot as plt
vectors = [[2,1],[1,3],[3,4]]
cone = PyNormaliz.NmzCone("cone",vectors)
HB = PyNormaliz.NmzResult(cone,"HilbertBasis")
for v in vectors:
    plt.plot([0,v[0]],[0,v[1]],'k-')
plt.show()
```
